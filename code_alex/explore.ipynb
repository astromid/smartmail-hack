```python
%%time
!mogrify -format jpg data/train/*/*
import glob
import os
paths = glob.glob('data/train/*/*')
gifs = [x.split('.')[0] for x in glob.glob('data/train/*/*.gif')]
print('All -> jpg')
for path in paths:
    resol = path.split('.')[-1]
    if resol != 'jpg':
        os.remove(path)
        #print('Image %s removed' % path)
        
new_paths = glob.glob('data/train/*/*')
for path in new_paths:
    if path.split('-')[0] in gifs:
        if path[-6:] != '-0.jpg':
            os.remove(path)
            #print(path)
print('Done!')
```
